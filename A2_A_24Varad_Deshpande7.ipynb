**Theory :**

Probability distributions describe how values of a random variable are distributed. They are fundamental in statistics, data analysis, and machine learning for modeling uncertainty and predicting outcomes.

Binomial Distribution models the number of successes in a fixed number of independent trials with only two possible outcomes (success or failure).
Poisson Distribution models the number of events occurring in a fixed interval of time or space when events occur independently at a constant rate.
Normal Distribution is a continuous distribution that is symmetric and bell-shaped, representing many natural phenomena such as height, weight, and measurement errors.

These distributions help in understanding patterns, making predictions, and building statistical models.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

n = 12
p = 0.4

binomial_samples = binom.rvs(n, p, size=1000)

plt.figure(figsize=(6,4))
plt.hist(binomial_samples, bins=range(0, n+2), density=True, alpha=0.7)
plt.title("Binomial Distribution (Histogram View)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

lam = 4

x = np.arange(0, 15)
poisson_values = poisson.pmf(x, lam)

plt.figure(figsize=(6,4))
plt.scatter(x, poisson_values)
plt.plot(x, poisson_values, linestyle='--')
plt.title("Poisson Distribution (Scatter + Line)")
plt.xlabel("Event Count")
plt.ylabel("Probability")
plt.grid(True)
plt.show()

mean = 2
std_dev = 1.5

x = np.linspace(-5, 10, 200)
normal_curve = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(6,4))
plt.plot(x, normal_curve)
plt.fill_between(x, normal_curve, alpha=0.3)
plt.title("Normal Distribution (Filled Curve)")
plt.xlabel("Values")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

**1. Difference between Binomial, Poisson, and Normal Distributions**

The three distributions differ based on type, application, and characteristics:

1. Binomial Distribution is discrete and deals with a fixed number of trials. Each
trial has only two outcomes, and the probability of success remains constant. Example: number of heads in coin tosses.

2. Poisson Distribution is also discrete but focuses on counting the number of events in a fixed interval. It does not require a fixed number of trials but assumes events occur randomly and independently. Example: number of calls received in an hour.

3. Normal Distribution is continuous and represents data that is symmetrically distributed around the mean. It is defined by mean and standard deviation. Example: distribution of student heights.

**2. Real-World Significance of Poisson Distribution**

The Poisson distribution is important for modeling rare and random events in real life.

* It helps in predicting how many times an event will occur in a given period.

* It is widely used in areas like:

        1.Call centers (number of calls per minute)

        2.Traffic flow (number of cars passing a signal)

        3.Healthcare (number of patients arriving at a hospital)

        4.Network systems (number of server requests)

**3. Central Limit Theorem (CLT) and its Relation to Normal Distribution**

The Central Limit Theorem (CLT) states that:
When the sample size is large enough, the sampling distribution of the sample mean tends to follow a Normal distribution, regardless of the original population distribution.


Explanation:

* Even if data is skewed or irregular, taking many samples and computing their means will produce a distribution that is approximately normal.
This is why the Normal distribution appears so frequently in statistics.


* Relation to Normal Distribution:


* CLT explains why Normal distribution is universal.
It allows us to use Normal distribution techniques even when the original data is not normally distributed.

**4. Why Normal Distribution is Widely Used in Machine Learning and Statistics**

The Normal distribution is widely used due to several important properties:

* It is mathematically simple and easy to work with.

* Many real-world data sets naturally follow a Normal distribution.

* It is defined by only two parameters: mean and standard deviation.

* It supports powerful statistical methods like hypothesis testing and confidence intervals.

* Many machine learning algorithms assume normality (e.g., Naive Bayes, Linear Regression).

**5. Skewness in Distributions and Which Distributions Exhibit It**

Skewness refers to the asymmetry of a probability distribution.

1. Positive Skew (Right Skew): Tail extends to the right (mean > median).
2. Negative Skew (Left Skew): Tail extends to the left (mean < median).
3. Zero Skew: Perfect symmetry.

* Distribution-wise behavior:

      Binomial Distribution:
          Can be symmetric or skewed depending on probability (p).
            p = 0.5 → symmetric
            p ≠ 0.5 → skewed

      Poisson Distribution:
          Usually positively skewed, especially when the average rate (λ) is small.

      Normal Distribution:
          Always perfectly symmetric (zero skewness).
