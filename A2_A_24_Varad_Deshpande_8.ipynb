## 1. What is correlation? How is it different from regression?

**Correlation** measures the strength and direction of a linear relationship between two variables. It is expressed by a coefficient (usually Pearson’s $r$) ranging from -1 to +1.

**Regression** goes a step further by describing the *functional* relationship between variables. It allows you to predict the value of a dependent variable ($Y$) based on the value of an independent variable ($X$).

| Feature | Correlation | Regression |
| :--- | :--- | :--- |
| **Purpose** | To find a numerical value expressing the relationship. | To explain the impact of change and predict future values. |
| **Symmetry** | Symmetric: Correlation between $X$ and $Y$ is the same as $Y$ and $X$. | Asymmetric: Predicting $Y$ from $X$ is different from predicting $X$ from $Y$. |
| **Output** | A single point (coefficient). | An equation (a line or curve). |
| **Cause/Effect** | Does not imply causation. | Often used to model a mathematical dependency. |

## 2. What is simple linear regression? Explain with suitable example.

**Simple Linear Regression (SLR)** is a statistical method that allows us to summarize and study relationships between two continuous variables. It models the relationship using a straight line.

The formula is:
$$Y = \beta_0 + \beta_1 X + \epsilon$$

*   $Y$: The Dependent Variable (what you want to predict).
*   $X$: The Independent Variable (the predictor).
*   $\beta_0$: The Y-intercept.
*   $\beta_1$: The Slope (the change in $Y$ for every 1 unit change in $X$).
*   $\epsilon$: The error term (residuals).

**Example:**
Imagine you are a coffee shop owner. You want to predict **daily revenue ($Y$)** based on the **outside temperature ($X$)**.
If your regression model is $Y = 50 + 20X$, it means that even if the temperature is 0, you expect to make $50. For every degree the temperature rises, your revenue increases by $20.

### 3. How do you calculate Pearson correlation in Python?
In the Python ecosystem, you typically use **pandas** for quick checks or **scipy** for more detailed statistical analysis.

**Using Pandas:**
This is the fastest way to get a single correlation value or a full matrix from your dataframe.
```python
# Returns a single correlation value between resolution and price
correlation = df['Max resolution'].corr(df['Price'])
print(f"Pearson Correlation: {correlation:.4f}")
```

**Using Scipy (to get the p-value):**
The `pearsonr` function from `scipy.stats` is preferred when you need to check the statistical significance (p-value). It is crucial to drop `NaN` (missing) values first to avoid errors.

```python
from scipy.stats import pearsonr

# 1. Clean the data (remove rows where either value is missing)
df_clean = df[['Max resolution', 'Price']].dropna()

# 2. Calculate the coefficient (r) and the p-value
corr_coeff, p_value = pearsonr(df_clean['Max resolution'], df_clean['Price'])

print(f"Karl Pearson Correlation Coefficient: {corr_coeff:.4f}")
print(f"P-value: {p_value:.4f}")
```

### Understanding the Result

*   **Correlation Coefficient ($r$):** This indicates a **weak positive relationship**, meaning as resolution increases, price tends to increase only slightly.
*   **P-value:** This tells you if the correlation is statistically significant. If it is less than 0.05, you can be confident the relationship is not due to random chance.

## 4. What is multicollinearity? Does it affect simple regression?

**Multicollinearity** occurs when two or more independent variables in a multiple regression model are highly correlated with each other. This is a headache because it makes it difficult for the model to determine which variable is actually doing the "heavy lifting" for the prediction.

**Does it affect Simple Regression?**
**No.** By definition, Simple Linear Regression involves only **one** independent variable. Since there isn't a second independent variable to be correlated with, multicollinearity cannot exist in a simple regression setup.

----

## 5. A dataset shows strong correlation but poor prediction accuracy. Why?

It sounds counterintuitive, but a high correlation coefficient doesn't always guarantee a perfect model. Here is why:

*   **Non-linearity:** Pearson correlation measures *linear* relationships. If your data follows a complex curve, the "strong" linear correlation might not capture the full picture, leading to a poor linear fit.
*   **Outliers:** A few extreme data points can pull the correlation coefficient higher (or lower) than it should be, while the rest of the data fails to follow the trend line closely.
*   **High Variance (Noise):** You might have a clear upward trend (high correlation), but if the data points are widely scattered around that trend line, the individual predictions will have high error.
*   **The "Hidden Factor":** You might have a high correlation between $X$ and $Y$, but $Y$ might actually be influenced by five other variables you aren't measuring. Without those, your model is "underfitted."

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Mount Google Drive to access the dataset
from google.colab import drive
drive.mount('/content/drive')

# Load the dataset
file_path = '/content/drive/My Drive/camera_dataset.csv'
df = pd.read_csv(file_path)

# --- Karl Pearson's Correlation Coefficient ---
print("\n--- Karl Pearson's Correlation Coefficient ---")
# Ensure 'Max resolution' and 'Price' columns are clean for correlation calculation
df_clean_corr = df[['Max resolution', 'Price']].dropna()

if not df_clean_corr.empty:
    corr_coeff, p_value = pearsonr(df_clean_corr['Max resolution'], df_clean_corr['Price'])
    print(f"Correlation between Max Resolution and Price: {corr_coeff:.4f}")
    print(f"P-value: {p_value:.4f}")
else:
    print("Not enough data after dropping NaNs to calculate correlation for Max resolution and Price.")


# --- Simple Linear Regression ---
print("\n--- Simple Linear Regression ---")
# For SLR, we also need to ensure the columns are clean
df_clean_slr = df[['Max resolution', 'Price']].dropna()

if not df_clean_slr.empty:
    X = df_clean_slr[['Max resolution']] # Independent variable (must be 2D)
    y = df_clean_slr['Price']             # Dependent variable

    # Initialize and train the model
    model = LinearRegression()
    model.fit(X, y)

    # Get model parameters
    slope = model.coef_[0]
    intercept = model.intercept_
    r_squared = r2_score(y, model.predict(X))

    print(f"Regression Equation: Price = ({slope:.4f} * Max resolution) + {intercept:.4f}")
    print(f"R-squared Value: {r_squared:.4f}")
else:
    print("Not enough data after dropping NaNs to perform Simple Linear Regression for Max resolution and Price.")
````

## **Graph: Simple Linear Regression - Max Resolution vs. Price**

````python
# Ensure df_clean_slr is available from the previous cell's execution
# If the previous cell was skipped or failed, ensure this block handles it
if 'df_clean_slr' in locals() and not df_clean_slr.empty:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Max resolution', y='Price', data=df_clean_slr, label='Data Points', alpha=0.6)

    # Plot the regression line using the model from the previous cell
    if 'model' in locals():
        plt.plot(df_clean_slr['Max resolution'], model.predict(df_clean_slr[['Max resolution']]),
                 color='red', linewidth=2, label='Regression Line')

    plt.title('Simple Linear Regression: Max Resolution vs. Price')
    plt.xlabel('Max Resolution (megapixels)')
    plt.ylabel('Price ($)')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
else:
    print("Cannot generate graph: Data for Simple Linear Regression is not available or empty.")
````
