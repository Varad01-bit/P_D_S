**Aim**: Data Ingestion and Cleaning: Acquire data and prepare it for analysis.


**Theory**:

Data ingestion is the process of collecting and importing data from different sources such as CSV files, databases, or APIs into a system for analysis. In this assignment, the Titanic dataset was loaded using the Pandas library with the read_csv() function, which converts raw data into a structured DataFrame. Functions like head() and tail() were used to view the first and last few rows of the dataset, helping to understand its structure and contents.


Data cleaning is the process of identifying and correcting errors or inconsistencies in the dataset. Real-world data often contains missing values, duplicates, or incorrect formats. In this dataset, missing values were found in columns such as Age, Cabin, and Embarked. These were identified using functions like isnull() and sum(). Cleaning the data is important to improve accuracy, ensure consistency, and make the dataset suitable for analysis and machine learning tasks.


Together, data ingestion and cleaning form the foundation of data preprocessing, ensuring that raw data is transformed into a reliable and usable format.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv" #dataset
df = pd.read_csv(url)

print("Data Loaded")
print("\nFirst 5 rows (head):")
print(df.head())
print("\nLast 5 rows (tail):")
print(df.tail())

missing_data = df.isnull().sum()

print("Missing values per column:")
print(missing_data)

total_cells = df.size
total_missing = missing_data.sum()
print(f"\nTotal missing values in dataset: {total_missing}")
```

**2. Why is data cleaning necessary?**

Definition: Data cleaning is the process of identifying and fixing errors in raw data.

Common issues:
1. Missing values (NaN or null entries)
2. Duplicate records
3. Incorrect data formats (e.g., dates as text)
4. Outliers or invalid values


Why it is important:

Data cleaning improves the accuracy of analysis, ensures reliable machine learning models, and prevents misleading results. It involves several common techniques such as removing or filling missing values, dropping duplicate records, converting data types into the correct format, and standardizing data to maintain consistency across the dataset.

Tools:
Libraries like Pandas are widely used for cleaning data efficiently.

**3. How do you load a CSV file into a Pandas DataFrame?**

To load a CSV file into a Pandas DataFrame, you use the read_csv() function from Pandas. First, import Pandas, then provide the file path to the function. It reads the CSV file and converts it into a DataFrame for analysis.

**What is the difference between head() and tail() functions in Pandas?**

**head() in Pandas:**
Returns the first n rows of a DataFrame (default is 5).

Syntax: df.head(n)

If n is not specified, it shows the first 5 rows.

Useful for:
1. Quickly previewing the dataset
2. Checking column names and structure
3. *Understanding* initial data entries

**tail() in Pandas:**
Returns the last n rows of a DataFrame (default is 5).

Syntax: df.tail(n)

If n is not specified, it shows the last 5 rows.

Useful for:
1. Viewing the end of the dataset
2. Checking if data is complete or truncated
3. Inspecting recent or last entries

**5. How can you identify missing values in a dataset?**

To identify missing values in a dataset using Pandas, you can use several built-in functions that help detect null or NaN values.

Common methods:
1. isnull() / isna()
Returns a DataFrame of Boolean values (True where data is missing).

2. sum() with isnull()
Shows the total number of missing values in each column.
