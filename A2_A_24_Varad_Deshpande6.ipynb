**Theory**:

Web scraping is the technique of automatically extracting data from websites using scripts and dedicated tools. In Python, two popular libraries are commonly employed for this purpose:

requests – used to send HTTP requests to a web page and retrieve its raw content.

BeautifulSoup (bs4) – used to parse the HTML structure and extract specific pieces of information.

Steps and Techniques in This Assignment
In this assignment, web scraping was carried out on a webpage containing tabular data. The process involved the following:

Request with custom headers – A User-Agent header (and optionally other headers) was added to the request to mimic a real browser, helping to avoid basic bot detection mechanisms.

Parsing and element location – After fetching the page content, BeautifulSoup was used to parse the HTML and locate target elements using tags, class names, and id attributes.

Handling missing or broken HTML tags – Special care was taken to prevent crashes. Conditional checks (e.g., testing if an element exists before accessing it) were implemented so that the script could still extract whatever data was available.

Output and Storage
The extracted data—especially from HTML tables—was processed into a structured format. Finally, the cleaned data was saved as a CSV file, making it readily available for further analysis. Libraries such as Pandas can also be used afterwards to organize and verify the scraped data.

Specific Tasks Performed in This Practical
Scrape tabular data from a webpage.

Store the scraped data in a CSV file.

Handle missing or broken HTML tags gracefully.

Extract data using class and id attributes.

Use the requests library with custom headers.

```python
!pip install beautifulSoup
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

print("Libraries imported successfully.")
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}
url = "https://en.wikipedia.org/wiki/List_of_brightest_stars"

def run_assignment_6():
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', class_='wikitable')

    scraped_rows = []

    if table:
        rows = table.find_all('tr')

        for row in rows:
            cells = row.find_all(['th', 'td'])
            cleaned_row = []
            for cell in cells:
                if cell:
                    cleaned_row.append(cell.get_text(strip=True))
                else:
                    cleaned_row.append("N/A")

            if cleaned_row:
                scraped_rows.append(cleaned_row)


        with open('stars_data.csv', 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerows(scraped_rows)

        print(f"Scrape Complete: {len(scraped_rows)} rows saved to stars_data.csv")
    else:
        print("Table target not found. Check the class attribute or URL.")

run_assignment_6()

try:
    df = pd.read_csv('stars_data.csv')
    print("Verification: First 10 rows of the stored data:")
    display(df.head(10))
except Exception as e:
    print(f"Verification Failed: {e}")
```
