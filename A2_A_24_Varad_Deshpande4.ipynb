NumPy (short for Numerical Python) is a foundational library for scientific and numerical computing in Python. It offers robust support for large, multi‑dimensional arrays and matrices, along with an extensive collection of mathematical functions designed for efficient, element‑wise operations.

In this assignment, NumPy arrays were created and manipulated to demonstrate core techniques such as indexing, slicing, and basic arithmetic computations. A two‑dimensional (2D) array was used to show how individual elements can be accessed via index positions, and how sub‑arrays (slices) can be extracted. NumPy’s design makes handling such structured data straightforward. Additionally, a one‑dimensional (1D) array was employed to perform common statistical operations—calculating squares, sums, and means—which are frequently encountered in data analysis.

Why is NumPy faster than pure Python?

It leverages vectorization, meaning operations are applied to entire arrays at once rather than looping over individual elements.

Its core routines are implemented in C, resulting in high execution speed.

Data is stored in contiguous memory blocks, which improves memory access patterns and computational efficiency.

NumPy supports a wide range of mathematical operations, including:

Arithmetic – addition, subtraction, multiplication, etc.

Statistical functions – mean, sum, standard deviation.

Advanced computations – linear algebra (matrix multiplication, determinants), trigonometric functions, and more.

Its powerful indexing and slicing capabilities allow users to access and modify specific portions of data without unnecessary copying.

Finally, NumPy serves as the underpinning for most other Python data science libraries, including:

Pandas (data wrangling and analysis)

Matplotlib (visualisation)

Scikit‑learn (machine learning)


```python
import numpy as np

arr_2d = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])

print("2D Array:\n", arr_2d)

print("\nElement at index [1, 2]:", arr_2d[1, 2]) # Output: 60

print("\nSliced Sub-array:\n", arr_2d[0:2, 1:3])

arr = np.array([1, 2, 3, 4])
print("\nOriginal Array:", arr)
print("Square of Array:", np.square(arr))
print("Sum of elements:", np.sum(arr))
print("Mean value:", np.mean(arr))
```

**2. Why is NumPy faster than Pure Python?**

NumPy is significantly faster for numerical operations because:

  1. Vectorization: It performs operations on entire arrays at once without needing explicit for loops in Python.

  2. C-Implementation: Its core functions are written in C, which allows for highly optimized, low-level execution.

  3. Contiguous Memory: NumPy arrays are stored in a continuous block of memory, making access much faster than Python lists.
