**Aim**:Data Wrangling with pandas, Manipulate and transform data to suit the analysis

**Theory:**

Data wrangling, also known as data munging, is the process of cleaning, transforming, and organizing raw data into a structured and usable format for analysis. In real-world scenarios, data is often incomplete, inconsistent, or unstructured, making wrangling an essential step before performing any analysis or building models.

In this assignment, a sample dataset was created using Pandas, which contained missing values in columns such as Name, Age, and Score. Functions like isnull() were used to identify missing values, helping to understand where cleaning was required. Missing values in the Age column were filled using the mean value with the help of fillna(), ensuring that the dataset remains complete without losing important records. Rows with missing Name values were removed using dropna() to maintain data consistency.

Data wrangling also includes transformation of data. A new column called Status was created based on the Score column using a conditional function. This process of creating new meaningful columns from existing data is known as column transformation and is an important part of feature engineering.

Additionally, Exploratory Data Analysis (EDA) plays a crucial role in understanding the dataset. It helps summarize data, identify patterns, detect outliers, and support better decision-making. Concepts like Series (1D), DataFrame (2D), and Panel (3D, now deprecated) represent different data structures in Pandas used for handling data efficiently

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values for demonstration
data = {
    'Name': ['Sarang', 'Amit', 'Rohit', 'Neha', None],
    'Age': [20, 22, np.nan, 21, 23],
    'Score': [85, np.nan, 90, 88, 75]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)


print("\nIs Null Check:")
print(df.isnull().sum())

# 4d: Fill missing Age with the mean Age (Data Wrangling step)
df['Age'] = df['Age'].fillna(df['Age'].mean())

# 4c: Drop rows where Name is missing
df = df.dropna(subset=['Name'])

# 5: Column Transformation (Example: Adding a 'Status' column)
# If score > 80, Status is 'Pass', else 'Average'
df['Status'] = df['Score'].apply(lambda x: 'Pass' if x > 80 else 'Average')

print("\nWrangled DataFrame:")
print(df)
```

**1. What is data wrangling?**

Data wrangling (or data munging) is the process of transforming and mapping data from one "raw" data form into another format with the intent of making it more appropriate and valuable for a variety of downstream purposes such as analytics.

**Data Wrangling:**

1. A broad process of cleaning, organizing, and transforming raw, messy data.
2. Focuses on making data consistent, usable, and well-structured.
3. Includes tasks like handling missing values, removing duplicates, merging datasets, and correcting formats.
4. Main goal is to prepare data for easy access and analysis.

**Data Preprocessing:**

1. A more specific step that usually comes after data wrangling.
2. Focuses on preparing data for machine learning models.
3. Includes techniques like scaling, normalization, encoding categorical variables, and feature selection.
4. Main goal is to convert data into a format suitable for algorithms.
<br>

**2. Importance of Exploratory Data Analysis (EDA):**

Exploratory Data Analysis (EDA):
EDA is the process of analyzing and summarizing a dataset to understand its main characteristics, often using statistical and visual methods before building any model.

1. Summarizes the main features of a dataset
2. Uses visual methods like charts and graphs
3. Helps discover patterns and relationships in data
4. Identifies anomalies, outliers, and errors
5. Checks assumptions before formal modeling
6. Supports better decision-making in analysis
7. Often performed using tools like Pandas

**3. Main Differences: Series, DataFrame, and Panel:**

  Series: A 1D labeled array capable of holding any data type.

  DataFrame: A 2D labeled data structure with columns of potentially different types (like a spreadsheet).

  Panel: A 3D container for data (Note: This is largely deprecated in modern Pandas in favor of MultiIndex DataFrames).

**4. Explain the use of:**

  a. isnull(): Detects missing values; returns a boolean object of the same shape.

  b. notnull(): Opposite of isnull(); returns True for non-missing values.

  c. dropna(): Removes missing values (rows or columns) from the dataset.

  d. fillna(): Fills missing values with a specific value or a calculated statistic (like mean).





**5. Column Transformation:**

Column transformation is the process of modifying existing columns or creating new columns from existing data in a dataset.

1. Involves changing values within existing columns or generating new ones
2. Helps in feature engineering by creating more meaningful variables
3. Improves model performance by better representing the problem
4. Can include operations like scaling, encoding, arithmetic operations, or applying functions
5. Useful for converting raw data into a more usable form for analysis or machine learning
